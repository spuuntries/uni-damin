## **Feature Engineering**

We first transform the unstructured data into interpretable formats.  
Here are the transformations done:

1. `dateAndViews`: Split into `views` and `date` via regex and parsed as their appropriate data types (float (technically int would work as well) and datetime).
2. `age`: We grab `date` and transform it into `age` via subtraction with the current date on a scale of days.
3. `likes`: To convert the likes, we first skip any instances with the `12345678901234567890123456789` artifact. We then interpret the number as int and multiply by the quantifier ("K" for 1000 and "M" for 1000000)
4. `subCount`: We do a similar transform to likes, just skipping any with no values.
5. `numCaps`: Count the number of capitalized letters in the title.
6. `numNonStand`: Count the number of non-standard (i.e., unicode outside of the ASCII range) letters in the title.
7. `clickbaity`: We define the degree of clickbaity-ness as the number of capitalized letters and non-standard letters in the title.
8. `clickbaityPercent`: The proportion of the clickbaity letters against the total length of the title.
9. `durationSecs`: We split the duration on the delimiter ":", parsing as int, then multiplying them to their respective quantities by the seconds scale.

We do a dropna() on a few of these to ensure that the resulting set is clean of NaNs.


```python
df["views"] = df["dateAndViews"].str.extract(r"(\d{1,3}(?:,\d{3})*(?:\.\d+)?) views")
df["date"] = df["dateAndViews"].str.extract(r"(\d{1,2} \w{3} \d{4})")
df["views"] = df["views"].str.replace(",", "").astype(float)
df["date"] = pd.to_datetime(df["date"], format="%d %b %Y")
df["age"] = (datetime.now() - df["date"]).dt.days
df.drop("dateAndViews", axis=1, inplace=True)
df.dropna(inplace=True)
```

```python
def convert_likes(likes):
    likes = likes.replace("Like", "")
    if "12345678901234567890123456789" in likes or not likes:
        return np.nan
    else:
        if "K" in likes:
            return int(float(likes.replace("K", "")) * 1000)
        elif "M" in likes:
            return int(float(likes.replace("M", "")) * 1000000)
        else:
            return int(likes)


df["likes"] = df["likes"].apply(convert_likes)
df.dropna(inplace=True)
```

```python
def convert_subs(subs):
    subs = subs.replace("subscribers", "")
    if not subs:
        return np.nan
    else:
        if "K" in subs:
            return int(float(subs.replace("K", "")) * 1000)
        elif "M" in subs:
            return int(float(subs.replace("M", "")) * 1000000)
        else:
            return int(subs)


df["subCount"] = df["subCount"].apply(convert_subs)
```

```python
df["numCaps"] = df["title"].str.findall(r"[A-Z]").str.len()
```

```python
df["numNonStand"] = df["title"].str.findall(r"[^\x00-\x7F]").str.len()
```

```python
df["clickbaity"] = df["numCaps"] + df["numNonStand"]
df["clickbaity"] = df["clickbaity"][df["clickbaity"] > 0]
df["clickbaityPercent"] = df["clickbaity"] / df["title"].str.len() * 100
```

```python
def convert_dur(dur):
    dur = dur.split(":")
    if len(dur) == 3:
        return int(dur[0]) * 3600 + int(dur[1]) * 60 + int(dur[2])
    elif len(dur) == 2:
        return int(dur[0]) * 60 + int(dur[1])
    else:
        return int(dur[0])


df["durationSecs"] = df["duration"].apply(convert_dur)
```

```python
df.dropna(inplace=True)
```

## **Normalization**

We do a bit of a normalization by removing some data based on certain criteria:

1. `views`: We clamp the deviation down by masking only data with less that 4 standard deviation and higher than 25000.
2. `clickbaityPercent`: We clamp this down at <2 standard deviation.
3. `subCount`: We do the same as clickbaityPercent.


```python
df["views"] = df["views"][(np.abs(stats.zscore(df["views"])) < 4)]
df["views"] = df["views"][df["views"] > 25000]
# df["likes"] = df["likes"][(np.abs(stats.zscore(df["likes"])) < 0.5)]
# df["likes"] = df["likes"][df["likes"] > 10000]
df["clickbaityPercent"] = df["clickbaityPercent"][
    (np.abs(stats.zscore(df["clickbaityPercent"])) < 2)
]
df["subCount"] = df["subCount"][(np.abs(stats.zscore(df["subCount"])) < 2)]
# df["durationSecs"] = df["durationSecs"][(np.abs(stats.zscore(df["durationSecs"])) < 2)]
```

## **Correlation Matrix**

For the correlation matrix, we use a spearman rank coefficient. This is due to the fact that it's known to be a bit more robust compared to the standard peason coefficient, and testing seems to show that this coefficient reflects the data's relationships better.

We use this to look for possible relationships based on correlation. However, this does not imply causal relationships, we will still judge and weigh the pairings based on domain knowledge.


```python
corr_matrix = df[
    ["likes", "views", "subCount", "durationSecs", "clickbaityPercent", "age"]
].corr("spearman")

fig = px.imshow(
    corr_matrix,
    x=corr_matrix.columns,
    y=corr_matrix.columns,
    color_continuous_scale="plotly3",
    title="Correlation Matrix",
    labels=dict(color="Correlation"),
    text_auto=".2f",
)

fig.update_layout(
    title_x=0.5,
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    yaxis_autorange="reversed",
)

fig.show()
```

```python
corr_matrix = df[
    ["likes", "views", "subCount", "durationSecs", "clickbaityPercent", "age"]
].corr()

fig = px.imshow(
    corr_matrix,
    x=corr_matrix.columns,
    y=corr_matrix.columns,
    color_continuous_scale="plotly3",
    title="Correlation Matrix",
    labels=dict(color="Correlation"),
    text_auto=".2f",
)

fig.update_layout(
    title_x=0.5,
    xaxis_showgrid=False,
    yaxis_showgrid=False,
    yaxis_autorange="reversed",
)

fig.show()
```

## **Bar Plotting** (univariate)

Here, we plot on a univariate basis, this is to analyze the nature of skew within the data. (Although this will still be explored within the scatters)

A majority of the data seems to exhibit a heavy left weight skew, with a very heavy tail on the righthand side. While this does somewhat violate common assumptions made by most statistical techniques (namely that data should exhibit a gaussian distribution), this does follow the intuition of domain knowledge.

This is due to the fact that data that possibly exhibits value classes (as videos at popularity classes do) tend to follow a power law, specifically here, Zipf's law. That is, the quantity of the class will be inversely proportional to its value (e.g., 0-19k likes will have a higher proportionality to 140k-159k).

Because of this fit to the domain knowledge, we have decided against log normalizing the data. As we believe that, if we log the scale, there may be some contextuality that's lost. Perhaps log norm could be done on a preprocessing step prior to the modelling, however for preliminary EDA we believe that it's better to not possibly "muddy the waters" and work with domain knowledge more.


```python
import plotly.subplots as sp

numeric_columns = df.select_dtypes(include=["int64", "float64"]).columns

for col in numeric_columns:
    print(col)
    print("Skew:", round(df[col].skew(), 2))
    fig = sp.make_subplots(rows=1, cols=2, subplot_titles=("Histogram", "Box Plot"))
    trace1 = go.Histogram(x=df[col], name="Histogram", nbinsx=40)
    fig.add_trace(trace1, row=1, col=1)
    trace2 = go.Box(y=df[col], name="Box Plot", boxpoints="outliers")
    fig.add_trace(trace2, row=1, col=2)
    fig.update_layout(showlegend=False, title_text=f"Distribution of {col}")
    fig.show()
```

### **views v. subCount v. clickbaityPercent**

Here we take the comparison of views against the clickbaity-ness of the video, along with the subscriber count of the channel. We plot them into pairs onto a 2D scatter, and grouped the three into one as a 3D scatter.

This is based on our hypothesis that, by intuitive reasoning based on domain knowledge, the clickbaity-ness of a video must have an effect on the number of views that a video gains. We add the `subCount` by the domain knowledge that, as users gain more subscribers, the likelihood of them using more "clickbaity" techniques gets higher and higher in order to meet the demands of the market. This is also to determine whether a higher subscriber count would result in a higher view count yield, as intuitive reasoning based on domain knowledge would indicate.

Some insights:

- `subCount v. views` seems to indicate that, while the domain knowledge would dictate that views and subscriber counts directly correlate, the correlation may not be as significantly reflected within the resulting data. This is seen in the low $R^2$ of $\approx103\times10^{-3}$, which means that only about 10.3% of the data can be explained by this relationship. This suggests that subscriber count alone is not enough to guarantee a video's success and therefore depend on other factors and characteristics of the video in order to succeed.
- `clickbaityPercent v. views` seems to suggest the relationship of the amount of clickbait used in the titles may not entirely correlate as significantly as domain knowledge seems to indicate. In fact, there's a very minor downtrend ($R^2 \approx2\times10^{-3}$) in the view count, with the distribution exhibiting a right skew (i.e., the weight distribution clustering on the lefthand side, with a heavy tail to the right), as clickbaity-ness increases in the title. This could indicate a diminishing returns trend with an optimal maxima in the data.
- `clickbaityPercent v. subCount` shows that, as seen in plot 2, there seems to be a local optimal maxima which the data seems to converge on, with a very minor downtrend ($R^2\approx4\times10^{-3}$) that may indicate a diminishing returns trend.
- `views v. subCount v. clickbaityPercent` seems to confirm that hypothesis in plot 2 and 3 where there exists an optimal maxima within the data with respect to the clickbaity-ness of a title. Interestingly, this optima seems to exist for both comparison axes, which could suggest that going above the upper bound of this range yields diminishing returns for the video's performance.

### **durationSecs v. age v. clickbaityPercent**

Another possibility is that, more recent videos, in order to be more engaging and stimulating, may have had to resort to shorter content with more stimulation packed in.

Some insights:

- `durationSecs v. age` seems to indicate that, as expected from the domain knowledge, there exists a clumping behaviour in recent years which correlate to the shortening of the videos' durations. What is interesting however, is how the upwards trend as age increases isn't as significant as we had initially hoped, only reaching an $R^2\approx21\times10^{-3}$. This could indicate that either we have not enough representative data, which makes some sense given the context of the data as this is curated from my (Faiz) feed, or that the duration may have not as much decreased in recent years and that the older videos are just outliers. We believe that the reason may be a combination of both of these.

The overall trend stayed relatively stagnant, with only $R^2\approx1\times10^{-3}$. Combined from insights from the previous section above, we believe this stagnation is possibly due to the observed maxima, such that even in older videos, there was little to no incentive to increase the clickbaity-ness of a video.
